```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo o df
df = pd.read_csv('/content/imigrantes_canada.csv')
df
```

```python
# Verificando o df
df.info()
```

```python
# Criando um index
df.set_index('País', inplace = True)
```

```python
# Criando uma lista utilizando o metodo map
anos = list(map(str, range(1980, 2014))) # Utilizando os parametros de 1980 a 2014
anos
```

```python
# Extraindo os dados de uma linha
brasil = df.loc['Brasil', anos]
brasil
```

```python
# Fazendo uma comparação entre dois dados
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
# Fazendo um grafico com a comparacao entre os dados
plt.plot(df_comparacao['Brasil'], label = 'Brasil') # Plotando a informação de um dado
plt.plot(df_comparacao['Argentina'], label = 'Argentina') # Plotando a informação de um dado
plt.title('Comparação entre Brasil e Argentina') # Definindo o titulo
plt.xlabel('Ano') # Rotulanod eixo x
plt.ylabel('Número de Imigrantes') # Rotulando eixo y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # Definiindo a posicao dos ticks no eixo X
plt.legend() # Adicionando legendo ao gráfico
plt.show() # Exibindo o gráfico
```
